```python
# extract some new features
California_housing_prices['bedrooms_ratio']=California_housing_prices['total_bedrooms']/California_housing_prices['total_rooms']
California_housing_prices['household_rooms']=California_housing_prices['households']/California_housing_prices['total_rooms']
California_housing_prices['household_bed_rooms']=California_housing_prices['households']/California_housing_prices['total_bedrooms']
California_housing_prices['population_room_ratio']=California_housing_prices['population']/California_housing_prices['total_rooms']
California_housing_prices['population_bedroom_ratio']=California_housing_prices['population']/California_housing_prices['total_bedrooms']
California_housing_prices['population denisty']=California_housing_prices['population']/California_housing_prices['households']
```

#Regression model

## standerdization

```python
y=California_housing_prices['median_house_value']
x=California_housing_prices.drop(columns=['median_house_value'],axis=1)
```

```python
x.info()
```

```python
x_numeric=x[['latitude', 'housing_median_age', 'total_rooms', 'median_income',
       'household_rooms', 'household_bed_rooms', 'population denisty']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_numeric_std = scaler.fit_transform(x_numeric)
```

```python
x_numeric_std
```

```python
x_scaled=x.drop(columns=['latitude', 'housing_median_age', 'total_rooms', 'median_income',
       'household_rooms', 'household_bed_rooms', 'population denisty'])
x_scaled.info()
```

```python
x_scaled=x_scaled.join(pd.DataFrame(x_numeric_std,columns=['latitude', 'housing_median_age', 'total_rooms', 'median_income',
       'household_rooms', 'household_bed_rooms', 'population denisty']))
```

```python
x_scaled.head()
```

```python
x_scaled=x_scaled.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42,shuffle=True)
```

```python
x_train
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,y_predict)
```

```python
MSE
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_predict)
r2
```

## Random forest

```python
# baseline model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
rfr=RandomForestRegressor(random_state=42)
r2_score=cross_val_score(rfr,x_train,y_train,cv=5,scoring='r2')
neg_mse_score=cross_val_score(rfr,x_train,y_train,cv=5,scoring="neg_mean_squared_error")
mse_score=-neg_mse_score
print("="*10)
print(np.mean(r2_score))
print(np.mean(mse_score))
print("="*10)
rfr.fit(x_train,y_train)
rfr.score(x_test,y_test)
```
