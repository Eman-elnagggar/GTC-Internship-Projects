#Encode and split

- one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder()
encoded_meal = one_hot_encoder.fit_transform(Hotel_Booking_df[['meal']])
encoded_meal_df = pd.DataFrame(
    encoded_meal.toarray(),
    columns=one_hot_encoder.get_feature_names_out(['meal']),
    index=Hotel_Booking_df.index
)
Hotel_Booking_encoded_df = pd.concat([Hotel_Booking_df, encoded_meal_df], axis=1)
```

```python
one_hot_encoder = OneHotEncoder()
encoded_market = one_hot_encoder.fit_transform(Hotel_Booking_df[['market_segment']])
encoded_market_df = pd.DataFrame(
    encoded_market.toarray(),
    columns=one_hot_encoder.get_feature_names_out(['market_segment']),
    index=Hotel_Booking_df.index
)
Hotel_Booking_encoded_df = pd.concat([Hotel_Booking_df, encoded_market_df], axis=1)
```

- label encoding

```python
Hotel_Booking_encoded_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
Hotel_Booking_encoded_df['is_canceled'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['is_canceled'])
Hotel_Booking_encoded_df['distribution_channel'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['distribution_channel'])
Hotel_Booking_encoded_df['is_repeated_guest'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['is_repeated_guest'])
Hotel_Booking_encoded_df['reserved_room_type'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['reserved_room_type'])
Hotel_Booking_encoded_df['assigned_room_type'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['assigned_room_type'])
Hotel_Booking_encoded_df['deposit_type'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['deposit_type'])
Hotel_Booking_encoded_df['agent_id'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['agent_id'])
Hotel_Booking_encoded_df['company_id'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['company_id'])
Hotel_Booking_encoded_df['customer_type'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['customer_type'])
Hotel_Booking_encoded_df['is_family'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['is_family'])
Hotel_Booking_encoded_df['country'] = label_encoder.fit_transform(Hotel_Booking_encoded_df['country'])
```
